```python
import pandas as pd
```

```python
credit=pd.read_csv("tmdb_5000_credits.csv")
movie=pd.read_csv("tmdb_5000_movies.csv")
```

```python
movie.head(5)
```

```python
credit.head(5)
```

```python
movie=credit.merge(movie,on="title")
```

```python
short_movie=movie[["movie_id","title","overview","cast","crew","genres","keywords"]]
```

```python
short_movie.head(5)
```

```python
import ast
def filter(word):
    l=[]
    for i in ast.literal_eval(word):
        l.append(i["name"])
    return l
```

```python
short_movie["genres"]=short_movie["genres"].apply(filter)
```

```python
short_movie["keywords"]=short_movie["keywords"].apply(filter)
```

```python

def corret(word):
    l=[]
    counter=0
    for i in ast.literal_eval(word):
        if counter!=3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
short_movie["cast"]=short_movie["cast"].apply(corret)
```

```python
def corret2(word):
    l=[]
    for i in ast.literal_eval(word):
        if i["job"]=="Director":
            l.append(i["name"])
            
        
            break
    return l
```

```python
short_movie["crew"]=short_movie["crew"].apply(corret2)
```

```python
short_movie["overview"]=short_movie["overview"].astype(str)
```

```python
short_movie["overview"]=short_movie["overview"].apply(lambda x:x.split())
```

```python
short_movie["tags"]=short_movie["overview"]+short_movie["cast"]+short_movie["crew"]+short_movie["genres"]+short_movie["keywords"]
```

```python
final_movie=short_movie[["movie_id","title","tags"]]
```

```python
final_movie["tags"][0]
```

```python
final_movie["tags"]=final_movie["tags"].apply(lambda x:" ".join(x))
```

```python
final_movie["tags"][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Sam Worthington Zoe Saldana Sigourney Weaver James Cameron Action Adventure Fantasy Science Fiction culture clash future space war space colony society space travel futuristic romance space alien tribe alien planet cgi marine soldier battle love affair anti war power relations mind and soul 3d')
```

```python
final_movie["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vactor=cv.fit_transform(final_movie["tags"]).toarray()
```

```python
vactor
```

```python
cv.get_feature_names()
```

```python
# print(cv.vocabulary_)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vactor)
```

```python
def recommandaction(movie):
    movie_index=final_movie[final_movie["title"]==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda X:X[1])[1:20]

    for i in movie_list:
        print(final_movie.iloc[i[0]].title)
```

```python
recommandaction("Avengers: Age of Ultron")
```
